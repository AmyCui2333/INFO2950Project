```python
import pandas as pd
import numpy as np
apps = pd.read_csv("googleplaystore.csv",encoding='utf-8')
print(apps)
```

```python
apps.shape
```

```python
apps.info()
```

```python
apps = apps.drop(columns = ['Current Ver','Android Ver','Last Updated','Genres'])
apps.info()
```

```python
apps.isnull().sum()
```

```python
#clean out instances where fields are null
apps = apps[pd.notnull(apps['Rating'])]
apps = apps[pd.notnull(apps['Content Rating'])]
apps = apps[pd.notnull(apps['Type'])]
apps.isnull().sum()
```

```python
print(apps)
```

```python
#remove instances without a specific size
apps = apps[apps.Size != 'Varies with device']
```

```python
#transform necessary columns into numeric variables for further analysis
apps['Installs'] = apps['Installs'].apply(lambda x: x.replace('+','').replace(',',''))
apps['Size'] = apps['Size'].apply(lambda x: x.replace('M','').replace('k',''))
apps['Price'] = apps['Price'].apply(lambda x: x.replace('$',''))
apps['Installs'] = apps['Installs'].astype(int)
apps['Reviews'] = apps['Reviews'].astype(int)
apps['Size'] = apps['Size'].astype(float)
apps['Rating'] = apps['Rating'].astype(float)
apps['Price'] = apps['Price'].astype(float)
import math
apps['Installs'] = apps['Installs'].apply(lambda x: math.log(x,10))
apps['Reviews'] = apps['Reviews'].apply(lambda x: math.log(x))
```

```python
print(apps)
```

```python
apps = apps.rename(columns={'Content Rating': 'content_rating'})
writing = apps.to_csv(index=False)
```

```python
filename = "googleapps.csv"
f = open(filename,"w",encoding='utf-8-sig')
f.write(writing)
```
